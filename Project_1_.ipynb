```python
import mysql.connector
from faker import Faker
import random
import time

# Connect to MySQL
conn = mysql.connector.connect(
    host="localhost",
    user="root",              
    password="aravind_26",  
    database="placement_app"
)
cursor = conn.cursor()
```

```python
# Students Table
cursor.execute("""
CREATE TABLE IF NOT EXISTS students (
    student_id INT PRIMARY KEY,
    name VARCHAR(100),
    age INT,
    gender VARCHAR(10),
    email VARCHAR(100),
    phone VARCHAR(20),
    enrollment_year INT,
    course_batch VARCHAR(50),
    city VARCHAR(50),
    graduation_year INT
)
""")
conn.commit()
```

```python
import pandas as pd
from faker import Faker
import random

fake = Faker()
genders = ['Male', 'Female', 'Other']

records = []

for _ in range(500):
    enrollment_year = random.choice([2020, 2021, 2022])
    record = {
        'name': fake.name(),
        'age': random.randint(18, 25),
        'gender': random.choice(genders),
        'email': fake.email(),
        'phone': fake.phone_number(),
        'enrollment_year': enrollment_year,
        'course_batch': random.choice(['Batch A', 'Batch B', 'Batch C']),
        'city': fake.city(),
        'graduation_year': enrollment_year + 4
    }
    records.append(record)

# Convert to DataFrame
stu_data = pd.DataFrame(records)
```

```python
alter_query = """
    ALTER TABLE students
    MODIFY COLUMN student_id INT AUTO_INCREMENT 
"""

cursor.execute(alter_query)
conn.commit()

print("✅ student_id column modified to AUTO_INCREMENT PRIMARY KEY.")
```

```python
insert_query = """
    INSERT INTO students 
    (name, age, gender, email, phone, enrollment_year, course_batch, city, graduation_year)
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)
"""

for index, row in stu_data.iterrows():
    cursor.execute(insert_query, (
        row['name'],
        row['age'],
        row['gender'],
        row['email'],
        row['phone'],
        row['enrollment_year'],
        row['course_batch'],
        row['city'],
        row['graduation_year']
    ))

conn.commit()
print("✅ Student records inserted successfully.")
```

```python
create_query = """
CREATE TABLE IF NOT EXISTS programming (
    programming_id INT AUTO_INCREMENT PRIMARY KEY,
    student_id INT,
    language VARCHAR(50),
    problems_solved INT,
    assessments_completed INT,
    mini_projects INT,
    certifications_earned INT,
    latest_project_score INT,
    FOREIGN KEY (student_id) REFERENCES students(student_id)
)
"""

cursor.execute(create_query)
conn.commit()

print("✅ programming table created successfully.")
```

```python
programming_records = []

languages = ['Python', 'SQL']

for prog_id in range(1, 501):  # Matching your 500 students
    record = {
        'student_id': prog_id,
        'language': random.choice(languages),
        'problems_solved': random.randint(10, 100),
        'assessments_completed': random.randint(1, 10),
        'mini_projects': random.randint(0, 5),
        'certifications_earned': random.randint(0, 3),
        'latest_project_score': random.randint(50, 100)
    }
    programming_records.append(record)

programming_df = pd.DataFrame(programming_records)
```

```python
insert_query = """
    INSERT INTO programming 
    (student_id, language, problems_solved, assessments_completed, mini_projects, certifications_earned, latest_project_score)
    VALUES (%s, %s, %s, %s, %s, %s, %s)
"""

for index, row in programming_df.iterrows():
    cursor.execute(insert_query, (
        row['student_id'],
        row['language'],
        row['problems_solved'],
        row['assessments_completed'],
        row['mini_projects'],
        row['certifications_earned'],
        row['latest_project_score']
    ))

conn.commit()
print("✅ Programming records inserted successfully.")
```

```python
create_query = """
CREATE TABLE IF NOT EXISTS soft_skills (
    soft_skill_id INT AUTO_INCREMENT PRIMARY KEY,
    student_id INT,
    communication INT,
    teamwork INT,
    presentation INT,
    leadership INT,
    critical_thinking INT,
    interpersonal_skills INT,
    FOREIGN KEY (student_id) REFERENCES students(student_id)
)
"""

cursor.execute(create_query)
conn.commit()
print("✅ soft_skills table created successfully.")
```

```python
soft_skills_records = []

for sid in range(1, 501):  # 500 students
    record = {
        'student_id': sid,
        'communication': random.randint(60, 100),
        'teamwork': random.randint(60, 100),
        'presentation': random.randint(60, 100),
        'leadership': random.randint(60, 100),
        'critical_thinking': random.randint(60, 100),
        'interpersonal_skills': random.randint(60, 100)
    }
    soft_skills_records.append(record)

soft_df = pd.DataFrame(soft_skills_records)
```

```python
insert_query = """
INSERT INTO soft_skills
(student_id, communication, teamwork, presentation,
 leadership, critical_thinking, interpersonal_skills)
VALUES (%s, %s, %s, %s, %s, %s, %s)
"""

for _, row in soft_df.iterrows():
    cursor.execute(insert_query, (
        int(row['student_id']),
        int(row['communication']),
        int(row['teamwork']),
        int(row['presentation']),
        int(row['leadership']),
        int(row['critical_thinking']),
        int(row['interpersonal_skills'])
    ))

conn.commit()
print("✅ soft_skills records inserted successfully.")
```

```python
create_query = """
CREATE TABLE IF NOT EXISTS placements (
    placement_id INT AUTO_INCREMENT PRIMARY KEY,
    student_id INT,
    mock_interview_score INT,
    internships_completed INT,
    placement_status VARCHAR(20),
    company_name VARCHAR(100),
    placement_package INT,
    interview_rounds_cleared INT,
    placement_date DATE,
    FOREIGN KEY (student_id) REFERENCES students(student_id)
)
"""

cursor.execute(create_query)
conn.commit()
print("✅ placements table created successfully.")
```

```python
from datetime import datetime

statuses = ['Placed', 'Ready', 'Not Ready']
companies = ['Infosys', 'TCS', 'Wipro', 'Zoho', 'HCL', None]

placement_data = []

for sid in range(1, 501):
    status = random.choice(statuses)

    if status == 'Placed':
        company = random.choice([c for c in companies if c])
        package = random.randint(300000, 1000000)
        placement_date = fake.date_between(start_date='-1y', end_date='today')
    else:
        company = None
        package = 0
        placement_date = None

    record = {
        'student_id': sid,
        'mock_interview_score': random.randint(50, 100),
        'internships_completed': random.randint(0, 3),
        'placement_status': status,
        'company_name': company,
        'placement_package': package,
        'interview_rounds_cleared': random.randint(0, 5),
        'placement_date': placement_date
    }
    placement_data.append(record)

placement_df = pd.DataFrame(placement_data)
```

```python
insert_query = """
INSERT INTO placements 
(student_id, mock_interview_score, internships_completed, placement_status, company_name, placement_package, interview_rounds_cleared, placement_date)
VALUES (%s, %s, %s, %s, %s, %s, %s, %s)
"""

for index, row in placement_df.iterrows():
    cursor.execute(insert_query, (
        row['student_id'],
        row['mock_interview_score'],
        row['internships_completed'],
        row['placement_status'],
        row['company_name'],
        row['placement_package'],
        row['interview_rounds_cleared'],
        row['placement_date']
    ))

conn.commit()
print("✅ placements records inserted successfully.")
```

```python
 import warnings
import pandas as pd
import mysql.connector
from mysql.connector import Error

warnings.filterwarnings(
    "ignore",
    category=UserWarning,
    message=".*pandas only supports SQLAlchemy connectable.*"
)

conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="aravind_26",
    database="placement_app"
)

tables = ["students", "programming", "soft_skills", "placements"]

for tbl in tables:
    try:
        df = pd.read_sql_query(f"SELECT * FROM {tbl}", conn)
    except Error as e:
        print(f"❌ Cannot read table '{tbl}': {e}")
        continue
    # If successful, save to CSV
    df.to_csv(f"{tbl}_backup.csv", index=False)
    print(f"✅ Backed up '{tbl}' with {len(df)} rows")
```

```python
df3 = pd.read_sql_query(
    """
    SELECT s.student_id, s.name, ss.communication
    FROM students s
    JOIN soft_skills ss ON s.student_id = ss.student_id
    WHERE ss.communication > 8
    """,
    conn
)
df3
```

```python
df4 = pd.read_sql_query(
    """
    SELECT s.student_id, s.name, ss.leadership, ss.teamwork
    FROM students s
    JOIN soft_skills ss ON s.student_id = ss.student_id
    WHERE ss.leadership > 8 AND ss.teamwork > 8
    """,
    conn
)
df4
```

```python
df5 = pd.read_sql_query(
    """
    SELECT s.course_batch, AVG(pr.latest_project_score) AS avg_project_score
    FROM students s
    JOIN programming pr ON s.student_id = pr.student_id
    GROUP BY s.course_batch
    """,
    conn
)
df5
```

```python
df6 = pd.read_sql_query(
    "SELECT placement_status AS placed, COUNT(*) AS count FROM placements GROUP BY placement_status",
    conn
)
df6
```
